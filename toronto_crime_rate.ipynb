### Data Cleaning

Start by cleaning the data.

```python
#drop is a function that lets you delete an entire column.
#inplace = true means to replace it in the dataframe.
crimes.drop('X',axis=1, inplace=True)
crimes.drop('Y',axis=1, inplace=True)
```

```python
#see the columns that are stored in crimes
crimes.columns
```

Next, let's see what values are stored in the `offence` column. 

```python
#see the total number of offences (how many per category)
offence_counts = crimes.groupby('offence').size().reset_index(name = 'Offence counts').sort_values('Offence counts', ascending= False)
offence_counts.head(10)
```

Look at the Neighbourhood column. Notice how it has some extra numbers brackets and spaces at the end. Let's remove them.

```python
crimes['Neighbourhood']
```

```python
#Remove the characters at the end of the neighbourhood column values
crimes['Neighbourhood'] = crimes['Neighbourhood'].str[:-5]
```

Check how the data looks like.

```python
#look at data 
crimes.head(3)
```

Let's look at the years.

```python
#look at crimes column
crimes['occurrenceyear']
```

This is quite a lot to look at, so let's group them together and see the total per year.

```python
#group by the year and see how many in total per year
crimes.groupby('occurrenceyear').size().reset_index(name = 'Occurence years').sort_values('Occurence years', ascending= False)
```

Even thought there is data between 2000 and 2013, the numbers vary too much in comparison with 2014 to 2018. This might be due to lack of data collection during that period. For the sake of accuracy, we will only look at the interval of 2014-2018. 

```python
# remove anything that is before 2014
crimes = crimes[crimes['occurrenceyear'] > 2013]
```

```python
#rerun this line
crimes.groupby('occurrenceyear').size().reset_index(name = 'Occurence years').sort_values('Occurence years', ascending= False)
```

Save the dataframe in an external CSV for later use. (Not neccessary, but this is how one can do it).

```python
#save the modified dataframe to a csv to make more visualization plots in Tableau
crimes.to_csv(r'toronto_crime.csv',header=True)
```

We see that we have a lot of rows that are correlated to one another. For the purpose of this analysis, we will group them by categories. 

All `assault`- related crimes will be renamed to `Assault`.

```python
#change Assault-related crimes to "Assault"

crimes.offence = crimes.offence.replace({"Assault With Weapon":"Assault"}) 
crimes.offence = crimes.offence.replace({"Assault - Resist/ Prevent Seiz":"Assault"}) 
crimes.offence = crimes.offence.replace({"Aggravated Assault":"Assault"})
crimes.offence = crimes.offence.replace({"Assault - Force/Thrt/Impede":"Assault"})
crimes.offence = crimes.offence.replace({"Aggravated Assault Avails Pros":"Assault"})

#officers related assaults
crimes.offence = crimes.offence.replace({"Assault Peace Officer":"Assault"})
crimes.offence = crimes.offence.replace({"Aggravated Aslt Peace Officer":"Assault"})
crimes.offence = crimes.offence.replace({"Assault Peace Officer Wpn/Cbh":"Assault"})
crimes.offence = crimes.offence.replace({"Assault Bodily Harm":"Assault"})

#bodily harm involved
crimes.offence = crimes.offence.replace({"Unlawfully Causing Bodily Harm":"Assault"})
crimes.offence = crimes.offence.replace({"Crim Negligence Bodily Harm":"Assault"})
crimes.offence = crimes.offence.replace({"Traps Likely Cause Bodily Harm":"Assault"})
crimes.offence = crimes.offence.replace({"Set/Place Trap/Intend Death/Bh":"Assault"})
```

_Note_: according to definition, bodily harm means ... and assault means ... 

Next, there are a few crimes related to _Breaking and Entering_. We will rename those to `B&E`.

```python
#change 'B&E W'Intent' to 'B&E'
crimes.offence = crimes.offence.replace({"B&E W'Intent":"B&E"}) 
crimes.offence = crimes.offence.replace({"B&E Out":"B&E"}) 
crimes.offence = crimes.offence.replace({"B&E - M/Veh To Steal Firearm":"B&E"}) 
crimes.offence = crimes.offence.replace({"Unlawfully In Dwelling-House":"B&E"}) 
crimes.offence = crimes.offence.replace({"B&E - To Steal Firearm":"B&E"}) 
```

Next, change the _Robbery_-related crimes to `Robbery`

```python
# change Robbery With Weapon to Robbery
crimes.offence = crimes.offence.replace({"Robbery With Weapon":"Robbery"}) 
crimes.offence = crimes.offence.replace({"Robbery - Mugging":"Robbery"}) 
crimes.offence = crimes.offence.replace({"Robbery - Other":"Robbery"}) 
crimes.offence = crimes.offence.replace({"Robbery - Business":"Robbery"}) 
crimes.offence = crimes.offence.replace({"Robbery - Swarming":"Robbery"}) 
crimes.offence = crimes.offence.replace({"Robbery - Delivery Person":"Robbery"}) 
crimes.offence = crimes.offence.replace({"Robbery - Home Invasion":"Robbery"}) 
crimes.offence = crimes.offence.replace({"Robbery - Purse Snatch":"Robbery"}) 
crimes.offence = crimes.offence.replace({"Robbery - Financial Institute":"Robbery"}) 
crimes.offence = crimes.offence.replace({"Robbery - Vehicle Jacking":"Robbery"}) 
crimes.offence = crimes.offence.replace({"Robbery - Taxi":"Robbery"}) 
crimes.offence = crimes.offence.replace({"Robbery - Atm":"Robbery"})
crimes.offence = crimes.offence.replace({"Robbery - Armoured Car":"Robbery"})
```

Then change the _Theft_-related crimes to `Theft`

```python
# rename theft-related columns to theft
crimes.offence = crimes.offence.replace({"Theft Over - Distraction":"Theft"})
crimes.offence = crimes.offence.replace({"Theft Of Motor Vehicle":"Theft"}) 
crimes.offence = crimes.offence.replace({"Theft From Motor Vehicle Over":"Theft"}) 
crimes.offence = crimes.offence.replace({"Theft From Mail / Bag / Key":"Theft"}) 
crimes.offence = crimes.offence.replace({"Theft Over - Bicycle":"Theft"}) 
crimes.offence = crimes.offence.replace({"Theft Over - Shoplifting":"Theft"}) 
crimes.offence = crimes.offence.replace({"Theft - Misapprop Funds Over":"Theft"}) 
crimes.offence = crimes.offence.replace({"Theft Of Utilities Over":"Theft"}) 
crimes.offence = crimes.offence.replace({"Theft Over":"Theft"}) 
```

The _firearms_ related crimes will be grouped in `Firearms`.

```python
crimes.offence = crimes.offence.replace({"Discharge Firearm With Intent":"Firearms"}) 
crimes.offence = crimes.offence.replace({"Discharge Firearm - Recklessly":"Firearms"}) 
crimes.offence = crimes.offence.replace({"Pointing A Firearm":"Firearms"}) 
crimes.offence = crimes.offence.replace({"Use Firearm / Immit Commit Off":"Firearms"}) 
crimes.offence = crimes.offence.replace({"Air Gun Or Pistol: Bodily Harm":"Firearms"}) 

#disarming police officers
crimes.offence = crimes.offence.replace({"Disarming Peace/Public Officer":"Firearms"}) 
```

Now that we finished, let's take a look at the shape of the dataframe to make sure nothing was deleted.

```python
#checking the number of rows and columns to make sure nothing got deleted.
crimes.shape
```

Let's see how the table looks like after modifying it.

```python
#see the total number of offences (how many per category)
offence_counts = crimes.groupby('offence').size().reset_index(name = 'Offence counts').sort_values('Offence counts', ascending= False)
offence_counts
```

Let's take a look at the percentages of the the crimes.

```python
#calculate a percentage
top_offences = offence_counts['Offence counts']/len(crimes) * 100
#print it.
print(top_offences)
```

### Data Visualization

```python
#import pyplot library
from matplotlib import pyplot as plt
```

```python
#plot figure
plt.figure(figsize = (10,5))
plt.barh('offence', 'Offence counts', data = offence_counts)
plt.title("Crimes in Toronto")
plt.xlabel('Number of counts')
plt.show()
```

We can see in the above graphic that the highest crime in Toronto is `Assault`. When looking at the crime categories in Canada ([here]()), it is to notice that _push, pinch, tap_ are also considered to be assaults. While those can happen as accidents, they also happen on purpose.

```python
#pd.set_option('display.max_columns', None)
#pd.set_option('display.expand_frame_repr', False)
#pd.set_option('max_colwidth', -1)
```

What are the type of premises those crimes happen? 

```python
# see the premise types who are more likely for crimes
premise_counts = crimes.groupby('premisetype').size().reset_index(name = 'Premise counts').sort_values('Premise counts', ascending= False).head(10)
```

```python
#plot figure
plt.figure()
plt.bar('premisetype', 'Premise counts', data = premise_counts)
plt.title("Premises where crimes take places")
plt.xlabel('Number of counts')
plt.ylabel('Type of premise')
plt.show()
```

To no surprise, the most crimes happen outside. But the second highest are in apartments. That is honestly terrifying. 

What is the most popular hour for crimes? Is there a pattern we see?

```python
# what is the most popular hour
# see the premise types who are more likely for crimes
hour_counts = crimes.groupby('occurrencehour').size().reset_index(name = 'Occurence hour').sort_values('Occurence hour', ascending= False)
#see the top 5 
hour_counts.head(5)
```

```python
#plot figure
plt.figure(figsize = (20,10))
plt.bar('occurrencehour','Occurence hour', data = hour_counts)
plt.title("Hours most likely to have crimes")
plt.xlabel('Number of counts')
plt.yticks()
plt.show()
```

Surprisingly, the highest number of crimes occur at midnight and noon. It is interesting because they are opposite times of each other. The pattern we see is that most of them happen in the evening. 

The safest time to be out and about or inside the appartment is at 6 am.

## Data Exploration
Let's explore the data. See how it changes over time.

```python
#look at the nieghbourhoods with the highest number of crimes
crimes.groupby('Neighbourhood').size().reset_index(name = 'Numberofcrimes').sort_values('Numberofcrimes', ascending= False).head(10)
```

We see that those neighbourhoods have the highest crime rate. Let's see which crimes happen in those area.

```python
# See what crimes are done at Church-Yonge Corridor
top_first = crimes[crimes['Neighbourhood']== 'Church-Yonge Corridor'].groupby('offence').size().reset_index(name = 'Numberofcrimes').sort_values('Numberofcrimes', ascending= False)
top_first
```

```python
# see the crimes at Waterfront Communities - The Island
top_second = crimes[crimes['Neighbourhood']== 'Waterfront Communities-The Island'].groupby('offence').size().reset_index(name = 'Numberofcrimes').sort_values('Numberofcrimes', ascending= False)
top_second
```

```python
#See the crimes that take place at West Humber-Clairvill
top_third = crimes[crimes['Neighbourhood']== 'West Humber-Clairvill'].groupby('offence').size().reset_index(name = 'Numberofcrimes').sort_values('Numberofcrimes', ascending= False)
top_third
```

```python
#plot figure for Church-Yonge Corridor
plt.figure(figsize = (10,5))
plt.bar('offence','Numberofcrimes', data = top_first)
plt.title("Crime distribution that take place at Church-Yonge Corridor")
plt.xlabel('Number of counts')
plt.yticks()
plt.show()
```

```python
#plot figure for Waterfront Communities-The Island
plt.figure(figsize = (10,5))
plt.bar('offence','Numberofcrimes', data = top_second)
plt.title("Crime distribution that take place at Waterfront Communities-The Island")
plt.xlabel('Number of counts')
plt.yticks()
plt.show()
```

```python
#plot figure for West Humber-Clairvill
plt.figure(figsize = (10,5))
plt.bar('offence','Numberofcrimes', data = top_third)
plt.title("Crime distribution that take place at West Humber-Clairvill")
plt.xlabel('Number of counts')
plt.yticks()
plt.show()
```
